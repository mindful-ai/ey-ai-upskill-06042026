### Examples 

```python
classifier = pipeline("zero-shot-classification")
```

```python
classifier("This course is about Huggingface library and it's exploration", candidate_labels=["education", "politics", "business"],)
```

```python
classifier
```

```python
classifier = pipeline("ner", aggregation_strategy='simple')
print(classifier("SpaceX was founded by Elon Musk in the year 2017"))
```

### Using specific models

```python
generator = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M")
```

```python
data = generator("In this AI course, we will teach you how to: ", max_length=30, num_return_sequences=2)
```
