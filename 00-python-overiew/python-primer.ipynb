### 1.4 Write a script to determine if a number is prime or not.

```python
n = int(input("Enter a number: "))
for i in range(2, n):
    if n % i == 0:
        print(f"{n} is not a prime number")
        break
else: # this else is executed only if the loop is not broken - loop-else construct
    print(f"{n} is a prime number")
```

### 1.5 Functions

```python
def check_prime(n):
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
```

```python
check_prime(13)
```

```python
check_prime(15)
```

```python
n = int(input("Enter a number: "))
if check_prime(n):
    print(f"{n} is a prime number")
```

```python
primes = []
for n in range(2, 100):
    if check_prime(n):
        primes.append(n)
print(primes)
```

##### Comprehensions

```python
print([n for n in range(2, 100) if check_prime(n)]) # list comprehension
```
